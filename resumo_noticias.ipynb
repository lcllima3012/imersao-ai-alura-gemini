```python
#importar as dependências
import os
from google.colab import userdata
from google import genai
from google.genai import types
import requests

#criando a variável de ambiente e populando com a API Key gerada no Google AI Studio
os.environ['GOOGLE_API_KEY'] = userdata.get('GOOGLE_API_KEY')

#inicializa o cliente e guarda na variável client
client = genai.Client()

#definindo o modelo a ser usado
modelo = 'gemini-2.0-flash'

#criando a configuração da resposta - como o modelo deve responder à requisição
chat_config = types.GenerateContentConfig(
    system_instruction='Você resume textos recebidos de forma muito sucinta, extraindo apenas os principais pontos. \
    Ao receber textos com comandos HTML, você os ignora e foca somente no texto puro. \
    Ao fim do resumo, enumere os principais pontos que você encontrou no texto.',
)

#cria uma nova sessão do chat e guarda na variável chat
chat = client.chats.create(model=modelo, config=chat_config)

#função para extrair o texto de um link
def obter_texto_do_link(url):
  """
  Tenta obter o texto de um link usando a biblioteca requests.
  Retorna o texto se a requisição for bem-sucedida, caso contrário, retorna None.
  """
  try:
    resposta = requests.get(url)
    resposta.raise_for_status()  # Lança uma exceção para status de erro (4xx ou 5xx)
    return resposta.text
  except requests.exceptions.RequestException as e:
    print(f"Erro ao acessar o link: {e}")
    return None

# Link que você quer analisar
link_da_web = input('Informe o link da notícia ou digite "fim" para encerrar: ') # Substitua por um link real!

# Obtém o texto do link
texto_da_pagina = obter_texto_do_link(link_da_web)

# Verifica se o texto foi obtido com sucesso
if texto_da_pagina:
  print("Texto da página obtido com sucesso!")
else:
  print("Não foi possível obter o texto da página.")

#cria o loop para que o usuário insira o link da notícia
while texto_da_pagina != 'fim':
  response = chat.send_message(texto_da_pagina)
  print('Resumo: ', response.text)
  print('\n')
  print('\n')
  texto_da_pagina = input('Informe o link da notícia ou digite "fim" para encerrar: ')

```
